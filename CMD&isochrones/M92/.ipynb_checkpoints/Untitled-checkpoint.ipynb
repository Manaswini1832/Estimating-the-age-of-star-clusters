```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("M92_csv_data.csv",encoding="utf-8").fillna(0)
x = data['u-g'].iloc[0:].values
y = data['u'].iloc[0:].values
plt.xlabel("u-g")
plt.ylabel("u")
plt.title("CMD")
plt.ylim(16, 26)
plt.xlim(-2, 5)
plt.scatter(x,y,marker='+',zorder=2)
plt.savefig('M92_CMD.png')
plt.show()
```

```python
#Isochrone for age =14.2Gyr
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

u= np.loadtxt('Isochrone14.2_M92.txt',usecols=(5), unpack =True )
g= np.loadtxt('Isochrone14.2_M92.txt',usecols=(6), unpack =True )
diff = u-g
plt.plot(diff, u, 'k',zorder=1)
plt.xlabel('u-g')
plt.ylabel('u')
plt.savefig('Isochrone14.2_M92.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("M92_csv_data.csv",encoding="utf-8").fillna(0)
x = data['u-g'].iloc[0:].values
y = data['u'].iloc[0:].values
plt.xlabel("u-g")
plt.ylabel("u")
plt.title("Isochrone fitting")
# plt.ylim(35, 16)
plt.xlim(-2, 5)
plt.scatter(x,y-15,marker='+',zorder=2)

#Metallicity=-2.32, Age=1Gyr
u1= np.loadtxt('Isochrone1_M92.txt',usecols=(5), unpack =True )
g1= np.loadtxt('Isochrone1_M92.txt',usecols=(6), unpack =True )
diff1 = u1-g1
plt.plot(diff1, u1, 'y', zorder=1)

#Metallicity=-2.32, Age=5Gyr
u2= np.loadtxt('Isochrone5_M92.txt',usecols=(5), unpack =True )
g2= np.loadtxt('Isochrone5_M92.txt',usecols=(6), unpack =True )
diff2 = u2-g2
plt.plot(diff2, u2, 'c', zorder=1)

#Metallicity=-2.32, Age=9.5Gyr
u3= np.loadtxt('Isochrone9.5_M92.txt',usecols=(5), unpack =True )
g3= np.loadtxt('Isochrone9.5_M92.txt',usecols=(6), unpack =True )
diff3 = u3-g3
plt.plot(diff3, u3, 'r', zorder=1)

#Metallicity=-2.32, Age=14.2Gyr
u4= np.loadtxt('Isochrone14.2_M92.txt',usecols=(5), unpack =True )
g4= np.loadtxt('Isochrone14.2_M92.txt',usecols=(6), unpack =True )
diff4 = u4-g4
plt.plot(diff4, u4, 'k',zorder=1,label='Metallicity constant(-2.32)')
plt.xlabel('u-g')
plt.ylabel('u')
plt.legend(loc='upper left')
plt.savefig('Isochrone-fitted.png')
plt.show()
#black->14.2Gyr
#cyan->9.5Gyr
```
